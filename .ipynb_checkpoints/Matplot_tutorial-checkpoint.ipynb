# Maplotlib Tutorial

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
""" Use NumPy to generate random data

"""
import numpy as np

x = np.linspace(0, 10, 25)
y = x * x +2
print(x)
print(y)
```

```python
""" Only takes one command to draw
"""
pylab.plot(x,y, 'r') # 'r' stands for red
```

```python
""" Drawing a subgraph
"""
pylab.subplot(1,2,1) #The contents of the brackets represent (rows, columns, indexes)
pylab.plot(x,y, 'r--')
pylab.subplot(1,2, 2)
pylab.plot(y, x, 'g*-')
```

```python
#more

pylab.subplot(1,2,1) #The contents of the brackets represent (rows, columns, indexes)
pylab.plot(x,y, 'r--')
pylab.subplot(1,2, 1)
pylab.plot(y, x, 'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axes = fig.add_axes([0.5, 0.1, 0.8, 0.8])
axes.plot(x,y, 'r')
```

```python
#again we can draw subgraph
fig, axes = plt.subplots(nrows = 1, ncols=2)#submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
#we can also draw a picture or graph inside another graph
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])#big axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])#small axes

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
fig = plt.figure(figsize=(16,9),dpi=300)

fig.add_subplot()

plt.plot(x,y,'r')
```

```python
fig, axes = plt.subplots(figsize=(16,9),dpi=50)
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")
axes.set_title("Title")
axes.plot(x,y,'r')
```

```python
fig, axes = plt.subplots(dpi=200)
ax.legend(["label1","Label2"])
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")
axes.set_title("Title")
axes.plot(x,x**2)
axes.plot(x,x**3)
axes.legend(["y=x**2","y=x**3"],loc=2)
```

```python
# Line color, tranparency, and more
```

```python
fig, axes = plt.subplots(dpi=150)
axes.plot(x,x**2,color='red',alpha=.5)
axes.plot(x,x**.5,color = '#1155dd',alpha = .5)
axes.plot(x,x+3, color = 'green',alpha = .5)
```

```python
fig, ax = plt.subplots(dpi=150)

ax.plot(x,x+1, color='blue', lw=0.5, linestyle='-')
ax.plot(x,x+2, color='blue', linewidth=1, linestyle = '-.')
ax.plot(x,x+3, color='blue', linewidth=1.5, linestyle = ':')
#ax.plot(x,x+4, color='blue', linewidth=2)

line, = ax.plot(x,x+4, color='black', lw=1.5)
line.set_dashes([5,10,15,10])
```

```python
fig, ax = plt.subplots(dpi=150)

line, = ax.plot(x,x+1, color='black', lw=1.5)
line.set_dashes([15,15,15,15])
line, = ax.plot(x,x+2, color='black', lw=1.5)
line.set_dashes([5,2,15,10])
line, = ax.plot(x,x+3, color='green', lw=1.5)
line.set_dashes([5,10,15,13])
line, = ax.plot(x,x+4, color='blue', lw=1.5)
line.set_dashes([5,10,14,10])
```

```python
fig, ax = plt.subplots(dpi=150)

ax.plot(x,x+1, color='blue', lw=0.5, marker='o')
ax.plot(x,x+2, color='blue', linewidth=1, marker='+')
ax.plot(x,x+3, color='blue', linewidth=1.5, marker = 's')
ax.plot(x,x+4, color='blue', marker = '1')

ax.plot(x,x+5, color='blue', lw=0.5, marker='o',markersize=.5)
ax.plot(x,x+6, color='blue', lw=0.5, marker='o',markerfacecolor='red')
```

```python
fig, Axes = plt.subplots(1,2,figsize=(10,5))

Axes[0].plot(x, x**2,x, x**3,lw=2)
Axes[0].grid(True)

Axes[1].plot(x,x**2,x,x**3,lw=2)
Axes[1].grid(True)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
```

```python
n = np.array([0,1,2,3,4,5])
fig,axes = plt.subplots(1,4,figsize=(16,5))
axes[0].set_title("scatter plot")
axes[0].scatter(x, x+0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n, n**2,lw=2 )

axes[2].set_title("bar")
axes[2].bar(n, n**2,align="center", width=0.8, alpha=1)

axes[3].set_title("fill_between")
axes[3].fill_between(x, x**2,x**3, color='green',alpha=0.5)
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0,0.0,0.6,0.6], polar=True)

t= np.linspace(0, 2*np.pi,100)
ax.plot(t,t, color='blue',lw=3)
ax.plot(t,t*2, color='green',lw=3)
```

```python
n = np.random.randn(100000)
fig, axes = plt.subplots(1,2,figsize=(12,4))
axes[0].set_title("Default Histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative detailed historgram")
axes[1].hist(n, cumulative=True, bins=50)
```

```python
# Draw contour map

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0,3.0, delta)
y = np.arange(-2.0,2.0, delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 -(Y - 1)**2)
Z = (Z1 - Z2) *2

print(X)
print(Y)
print(Z)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline =1, fontsize=10)
ax.set_title("Contour Map")
```

```python
# Draw 3d surface images
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1,2,1, projection='3d')
ax.plot_surface(X,Y,Z, rstride=4,cstride=4,linewidth=0)
```

```python
fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(2,7,1, projection='3d')
ax.plot_surface(X,Y,Z, rstride=4,cstride=4,linewidth=0)
```

```python
fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1,1,1, projection='3d')
ax.plot_surface(X,Y,Z, rstride=4,cstride=4,linewidth=0)
```

```python
import matplotlib.pyplot as plt
#Data
languages = 'Python','Java','PHP','Javascript','C#','C++'
popularity = [22.2,17.6,8.8,8,7.7,6.7]
colors = ["#1f77b4","#ff7f0e","#2ca02c","#d62728","#9467bd","#8c564b"]
```

```python
explode = (0.1,0,0,0,0,0)
plt.pie(popularity, explode=explode,labels = languages,colors=colors,
        autopct='%1.1f%%',shadow=True,startangle=140)
```
